### SVD (LSA)

Singular value decomposition for Low-rank matrix approximation of orginal matrix


\begin{align}
A = U \Sigma V^T \approx \tilde{U} \tilde{\Sigma} \tilde{V}^T
\end{align}

by choosing n-largest singular values from in Sigam, we can have different approximarion of orginal matrix

```python
# Function for explained variance by different number of component in TruncatedSVD 
def explained_variance(data,n_components):
    svd = TruncatedSVD(n_components =n_components ,n_iter=7, random_state=42)
    svd.fit(data)
    return (svd.explained_variance_ratio_).sum()

```

```python
# Apply "explained_variance" function over different n_component to the smallest optimal n_component
n_components = list(range(2,107, 5)) # list of number of components
optimum_com = [explained_variance(X_train,n) for n in n_components]
optimum_com_tf = [explained_variance(X_tfidf,n) for n in n_components]
```

```python
#'According to graph TruncatedSVD of Frequncy matrix estimates more varince than TruncatedSVD of TF-IDF matrix at any n_components.'
# Plot shows that choosing count-matrix for SVD is outperforming TF-IDF matrix.

pairs = [optimum_com,n_components]
pairs_tf = [optimum_com_tf,n_components]


plt.plot(pairs[1], pairs[0], label = 'Count matrix')
plt.plot(pairs_tf[1], pairs_tf[0], label = 'TF_IDF matrix')

plt.title('Count matrix SVD vs TF-IDF SVD')
plt.ylabel("Explained variance ratio")
plt.xlabel(" Number of components")
plt.legend()
plt.show()

```

```python
# List of n_component with explained variance greater than 95%
# We will pick 82 as the n_components. Because, not only the approximated matrix is 95% similar to main matrix,
# but also it will result in 20 components fewer that than picking 102 component with almsot same rate of approximation with 96%
larger = [ (pairs[1][i],round(pairs[0][i], 3) )for i in range(len(pairs[0])) if pairs[0][i] > .95] 
larger
```

```python
# fit and transforming orginal data into truncated approximation
svd = TruncatedSVD(n_components =82 ,n_iter=7, random_state=42)
X_train_svd = svd.fit_transform(X_train)
X_test_svd = svd.transform(X_test)
X_train_svd.shape, X_test_svd.shape
```

# Model Development and Prediction


## 1.0 Logistic Regerssion

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn import metrics
from sklearn.metrics import classification_report

```

```python
y_test = y_test.ravel()
y_train = y_train.ravel()

y_test_ = np.where(y_test.ravel() == -1, 0,1)
y_train_ = np.where(y_train.ravel() == -1, 0,1)
```

## [Need more explanation]

```python
# "penalty='none'", mean logistic regression without regulizor.

# Fit
pipe = make_pipeline(StandardScaler(), LogisticRegression(penalty='none', max_iter= 5000, solver = 'saga')) 
pipe.fit(X_train_svd, y_train_)  # apply scaling on training data

# Predict
y_pred = pipe.predict(X_test_svd)

# # R^2
# print('Train Accuracy:',pipe.score(X_train_svd, y_train_))
# print('Test Accuracy:',pipe.score(X_test_svd, y_test_))

```

#### Model Evaluation using Confusion Matrix


```python
cnf_matrix = metrics.confusion_matrix(y_test_, y_pred)
cnf_matrix
```

```python
target_names = ['Spam', 'Non-spam']
print(classification_report(y_test_, y_pred, target_names=target_names))
```

#### ROC Curve


```python
#a plot of the true positive rate against the false positive rate. It shows the tradeoff between sensitivity and specificity.
y_pred_proba = pipe.predict_proba(X_test_svd)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test_,  y_pred_proba)
auc = metrics.roc_auc_score(y_test_, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

## Neural Network

```python
from keras.models import Sequential
from keras.layers.core import Dense
import tensorflow as tf
from scipy.stats import zscore
from keras.layers import Conv1D, MaxPool1D, GlobalMaxPooling1D, GlobalAveragePooling1D
from keras.layers import Dense, Embedding, GRU, LSTM, Dropout, Bidirectional
```

<b>NN structure:</b>
    
    ** 4 hidden layers with [80,60,40,20] units
    ** sigmoid output layer singel unit
    ** 82 input features

```python
# Fit
model3 = Sequential()
model3.add(Dense(80, activation="relu"))
model3.add(Dense(60, activation="relu"))
model3.add(Dense(40, activation="relu"))
model3.add(Dense(20, activation="relu"))
model3.add(Dense(1, activation="sigmoid")) # Since output is binary, we set Sigmoid as the activation function

# Since classification is binary, thus we put cross entropy as the Cost function, loss='binary_crossentropy'. 
model3.compile(loss='binary_crossentropy', optimizer="adam", metrics=['accuracy']) 
result3 = model3.fit(X_train_svd, y_train_,epochs=20, batch_size = 20, verbose=1)

result3
```

```python
model3.summary()
```

```python
# Predict
y_pred_nn = model3.predict(X_test_svd)
y_pred_nn
```
